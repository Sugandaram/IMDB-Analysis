```python
sns.set(style="whitegrid")
sns.boxplot(x=imdb1['duration'], boxprops = dict(facecolor = "lightblue"), medianprops = dict(color = "red", linewidth = 1.5))
```

```python
q1 = imdb1["duration"].quantile(0.25)
q3 = imdb1["duration"].quantile(0.75)
iqr = q3-q1
Ll = q1-1.5*iqr
Ul = q3+1.5*iqr
```

```python
Ll,Ul
```

```python
imdb1 = imdb1[(imdb1["duration"]>Ll) & (imdb1["duration"] < Ul)]
```

```python
imdb1["duration"].describe()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=imdb1['duration'], boxprops = dict(facecolor = "lightblue"), medianprops = dict(color = "green", linewidth = 1.5))
```
